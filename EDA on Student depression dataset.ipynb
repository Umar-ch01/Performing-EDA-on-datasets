```python
import pandas as pd

df = pd.read_csv('/content/Weather-related disease prediction.csv')
df.head()
```

```python
# Check data types and non-null values
print(df.info())

# Summary statistics
print(df.describe())

# Check for missing values in each column
print(df.isnull().sum())
```

```python
print(df.columns)
```

```python
# Numeric and Categorical Column Detection
num_cols = df.select_dtypes(include='number').columns.tolist()
cat_cols = df.select_dtypes(include='object').columns.tolist()

print("📊 Numeric Columns:", num_cols)
print("📋 Categorical Columns:", cat_cols)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
sns.boxplot(data=df[['Age', 'Temperature (C)', 'Humidity', 'Wind Speed (km/h)']])
plt.title("📦 Boxplot for Vital Signs")
plt.xticks(rotation=45)
plt.show()
```

```python
for col in ['Age', 'Temperature (C)', 'Humidity', 'Wind Speed (km/h)']:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, bins=30, color='steelblue')
    plt.title(f"📈 Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.countplot(data=df, x='prognosis', order=df['prognosis'].value_counts().index, palette='Set2')
plt.title("🧪 Prognosis Distribution")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
vitals = ['Age', 'Temperature (C)', 'Humidity', 'Wind Speed (km/h)']

plt.figure(figsize=(8, 5))
sns.heatmap(df[vitals].corr(), annot=True, cmap='coolwarm')
plt.title("🧪 Correlation Between Vital Signs")
plt.show()
```
